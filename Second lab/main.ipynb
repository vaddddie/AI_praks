```python
# 1.3.1

import numpy as np


arr = np.zeros([8, 8])

for count in range(2):
    for i in range(count, len(arr), 2):
        for j in range(count, len(arr), 2):
            arr[i, j] = 1
        
print(arr)
```

```python
# 1.3.4

import numpy as np


inp = input().split()
x = int(inp[0])
y = int(inp[1])

arr = np.zeros([x, y])

for index in range(x):
    arr[index][0] = 1
    arr[index][y - 1] = 1

for index in range(y):
    arr[0][index] = 1
    arr[x - 1][index] = 1

print(arr)
```

```python
# 2.3.1

import pandas as pd
import math


inpA = input().split()
inpB = input().split()

a = pd.Series({
    'x': int(inpA[0]),
    'y': int(inpA[1]),
})

b = pd.Series({
    'x': int(inpB[0]),
    'y': int(inpB[1]),
})

ans = math.sqrt((a['x'] - b['x']) ** 2 + (a['y'] - b['y']) ** 2)
print(ans)
```

```python
# 3.3.2

import pandas as pd
from sklearn import preprocessing as pr


link = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"

dataframe = pd.read_csv(link)

dataframe[["sepal_length_cm"]] = pr.MinMaxScaler().fit_transform(dataframe[["sepal_length_cm"]])
dataframe[["sepal_width_cm"]] = pr.StandardScaler().fit_transform(dataframe[["sepal_width_cm"]])
dataframe.head(5)
```
