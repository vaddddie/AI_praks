```python
# 1

def quality(x, y):
    return (x + 3 * y - 3)/(3 * x ** 2 + y ** 2 + 1)

def sorting(Z):
    sorted_id = sorted(range(len(Z)), key=lambda k: Z[k])

    return sorted_id

def exchangeScheme(oldX, oldY, sorted_id):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]

    X[0] = oldX[sorted_id[0]]
    X[1] = oldX[sorted_id[1]]
    X[2] = oldX[sorted_id[2]]
    X[3] = oldX[sorted_id[2]]

    Y[0] = oldY[sorted_id[2]]
    Y[1] = oldY[sorted_id[2]]
    Y[2] = oldY[sorted_id[0]]
    Y[3] = oldY[sorted_id[1]]

    return X, Y

def evo_step(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = X[:]

    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)

    return X, Y, Z

def evoSteps(X, Y, steps_num = 4):
    results = []

    for i in range(steps_num):
        quality_arr = [quality(x, Y[i]) for i, x in enumerate(X)]

        X, Y, Z = evo_step(X, Y, quality_arr)

        X, Y = exchangeScheme(X, Y, sorting(Z))

        results.append([X, Y, sum(quality_arr), quality_arr])

    return X, Y, results

X = [-5, -3, -2, 0]
Y = [-1, -2, 0, 1]

results = evoSteps(X, Y)

for i in range(len(results[2])):
    print(f"Max in {i + 1} step: {results[2][i][2]}")

quality_arr = []
for i in range(len(results[2])):
    quality_arr += results[2][i][3]

print(f"MAX quality: {max(quality_arr)}")
```

```python
# 2

from math import e
import networkx as nx

distances = [(1, 2, 24),
            (1, 3, 41),
            (1, 4, 36),
            (1, 5, 22),
            (1, 6, 19),
            (2, 3, 21),
            (2, 4, 33),
            (2, 5, 33),
            (2, 6, 14),
            (3, 4, 27),
            (3, 5, 39),
            (3, 6, 23),
            (4, 5, 20),
            (4, 6, 20),
            (5, 6, 19)]

path = [1, 3, 4, 5, 6, 2, 1]
swaps = [(3, 4),
        (4, 6),
        (5, 2),
        (6, 2)]
Random_numbers = [33, 82, 51, 76]

T = 100

display_graph(distances)
```

```python

def probability(delta, T):
    return 100 * e ** (-delta / T)

def reductTemp(prevT):
    return prevT * 0.5

def edgeLength(i, j, distances):
    return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1) for item in distances])

def routeLength(path, distances):
    edges = []

    for i in range(len(path) - 1):
        edges.append(edgeLength(path[i], path[i + 1], distances))

    return sum(edges)

def routeOneReplacement(path, swap):
    path[swap[0] - 1], path[swap[1] - 1] = path[swap[1] - 1], path[swap[0] - 1]
    return path

def chooseRoute(distances, path, swaps, T, Random_numbers):
    sumLength = routeLength(path, distances)
    sum_of_paths = [sumLength]

    for i in range(len(swaps)):
        new_way = routeOneReplacement(path[:], swaps[i])
        new_sum = routeLength(new_way, distances)
        sum_of_paths.append(new_sum)

        delta_sum = new_sum - sumLength

        if delta_sum > 0:
            prob = probability(delta_sum, T)

            if prob > Random_numbers[i]:
                path = new_way
                sumLength = new_sum

        else:
            path = new_way
            sumLength = new_sum

        T = reductTemp(T)

    return path, sum_of_paths

# ======================================================================================

def drawRouteGraph(distances, bestRoute):
    newDistances = []

    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    display_graph(newDistances)


def display_graph(distances):
    graph = nx.Graph()
    graph.add_weighted_edges_from(distances)
    nx.draw_kamada_kawai(graph, node_color="#fb7258", node_size=2000, with_labels=True)

# ======================================================================================

best_route, sum_of_paths = chooseRoute(distances, path, swaps, T, Random_numbers)

print(f"Best route: {best_route}")
print(f"Best route length: {routeLength(best_route, distances)}")
print(f"All lengths: {sum_of_paths}")

drawRouteGraph(distances, best_route)
```
