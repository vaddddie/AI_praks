```python
# 4.3.2

import numpy as np
import math
from matplotlib import pyplot as plt


def func(x):
    return (math.sqrt(1 + math.exp(math.sqrt(x)) + math.cos(x ** 2)) / abs(1 - math.sin(x) ** 2)) + math.log(abs(2 * x))


arr = np.arange(1, 11)

for i in range(len(arr)):
    arr[i] = func(arr[i])
    
fig = plt.figure(figsize=(10, 7))
    
plt.scatter(np.arange(0, math.floor(len(arr) / 2)), arr[:math.floor(len(arr) / 2)])

plt.plot(np.arange(math.floor(len(arr) / 2), len(arr)), arr[math.floor(len(arr) / 2):])

plt.grid()
plt.show()
```

```python
# 4.3.3
import numpy as np
import math
from matplotlib import pyplot as plt

def func(x):
    return abs(math.cos(x * math.exp(math.cos(x) + math.log(x + 1))))

x = np.arange(0, 11)
y = np.arange(0, 11.)

for i in range(len(y)):
    y[i] = func(y[i])

plt.plot(x, y, c='r')
plt.fill_between(x, y)
plt.grid()
plt.show()

print("S = ", np.trapz(x, y))
```

```python
# 4.3.4

import requests
from bs4 import BeautifulSoup
from random import choice
import numpy as np
from matplotlib import pyplot as plt

# =========================== Filling in the database ==========
companies = {
    "Apple": "https://finance.yahoo.com/quote/AAPL/history?period1=1609459200&period2=1640908800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true",
    "Google": "https://finance.yahoo.com/quote/GOOGL/history?period1=1609459200&period2=1640908800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true",
    "Microsoft": "https://finance.yahoo.com/quote/MSFT/history?period1=1609459200&period2=1640908800&interval=1mo&filter=history&frequency=1mo&includeAdjustedClose=true",
}

database = {
    "Apple": [],
    "Google": [],
    "Microsoft": [],
}

session = requests.Session()

users = open('Internet+Explorer.txt').read().split("\n")
userAgent = choice(users)

header = {
    "user-agent": userAgent
}

for company in companies:
    sait = session.get(companies[company], headers=header)
    html = BeautifulSoup(sait.content, "lxml")

    priceCell = html.find_all("tr", class_="BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)")

    for item in priceCell:
        tmp = item.find_next("td", class_="Py(10px) Pstart(10px)")
        if item.find("td", class_="Ta(start) Py(10px)") == None:
            database[company].insert(0, float(tmp.text))
            
# ================================================================
    
x = np.arange(1, 13)

plt.plot(x, database["Apple"], label="Apple", c='r')
plt.plot(x, database["Google"], label="Google", c='g')
plt.plot(x, database["Microsoft"], label="Microsoft", c='b')
plt.xlabel("Month")
plt.ylabel("Price")
plt.title("2021")
plt.legend()
plt.grid()
plt.show()
```

```python
# 4.3.5

import math


def Addition(x, y):
    return x + y

def Subtraction(x, y):
    return x - y

def Multiplication(x, y):
    return x * y

def Division(x, y):
    try:
        return x / y
    except:
        return "Operation error"

def Exponent(x, y):
    try:
        return math.exp(x + y)
    except:
        return "Operation error"
    
def Cosine(x, y):
    return math.cos(x + y)
    
def Sine(x, y):
    return math.sin(x + y)
    
def Exponentiation(x, y):
    try:
        return pow(x, y)
    except:
        return "Operation error"


inp = input("Enter x and y: ").split()
x = float(inp[0])
y = float(inp[1])

operations = {
    "+": Addition(x, y),
    "-": Subtraction(x, y),
    "*": Multiplication(x, y),
    "/": Division(x, y),
    "exp": Exponent(x, y),
    "cos": Cosine(x, y),
    "sin": Sine(x, y),
    "^": Exponentiation(x, y),
}

print("Choose an operation: ", [item for item in operations])
oper = input()

try:
    print(operations[oper])
except:
    print("Try again..")
```
