```python
# 1

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3])
y = np.array([-4, -0.4, 4.1, 4])

A = np.vstack([x, np.ones(len(x))]).T

m, c = np.linalg.lstsq(A, y, rcond = None)[0]

plt.plot(x, y, 'o', label='Initial data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Linear extrapolation')
plt.legend()
plt.show()
```

```python
from numpy import *
from numpy.random import *
import matplotlib.pyplot as plt

point_number = 150

delta = 1.0
x = linspace(-5, 5, point_number)
y = x**2 + delta * (rand(point_number) - 0.5) + 10 * x
x += delta * (rand(point_number) - 0.5)

m = vstack((x**2, x, ones(point_number))).T
s = linalg.lstsq(m, y, rcond=None)[0]

x_prec = linspace(-5, 5, 101)

plt.plot(x, y, 'D')
plt.plot(x_prec, s[0] * x_prec**2 + s[1] * x_prec + s[2], '-', lw=2)
plt.grid()
plt.show()
```

```python
from numpy import *
from numpy.random import *
import matplotlib.pyplot as plt

point_number = 150

delta = 0.5
x = linspace(-5, 5, point_number)
y = x**2 + delta * (rand(point_number) - 0.5) + 1 / (2 * x)
x += delta * (rand(point_number) - 0.5)

m = vstack((x**3, x**2, x, ones(point_number))).T
s = linalg.lstsq(m, y, rcond=None)[0]

x_prec = linspace(-5, 5, 101)

plt.plot(x, y, 'D')
plt.plot(x_prec, s[0] * x_prec**3 + s[1] * x_prec**2 + s[2] * x_prec + s[3], '-', lw=3)
plt.grid()
plt.show()
```

```python
# 2

import numpy as np
from scipy.optimize import curve_fit

b0 = 0.1
b1 = 0.7
b2 = 0.9

points_number = 150

def func(x, b0, b1, b2):
    return b0 + b1 * np.exp(-b2 * x**2)

beta = (b0, b1, b2)

xdata = np.linspace(0, 5, points_number)

y = func(xdata, *beta)
ydata = y + 0.05 * np.random.randn(len(xdata))

beta_opt, beta_cov = curve_fit(func, xdata, ydata)

lin_dev = sum(beta_cov[0])
print("Linear devation: ", lin_dev)

residuals = ydata - func(xdata, *beta_opt)
fres = sum(residuals**2)
print("Quadratic devation: ", lin_dev)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, func(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
# 2.1

import numpy as np
from scipy.optimize import curve_fit

b0 = 1.2
b1 = 0.2

points_number = 150

def func(x, b0, b1):
    return b0 + b1 * x

beta = (b0, b1)

xdata = np.linspace(0, 5, points_number)

y = func(xdata, *beta)
ydata = y + 0.05 * np.random.randn(len(xdata))

beta_opt, beta_cov = curve_fit(func, xdata, ydata)

lin_dev = sum(beta_cov[0])
print("Linear devation: ", lin_dev)

residuals = ydata - func(xdata, *beta_opt)
fres = sum(residuals**2)
print("Quadratic devation: ", lin_dev)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, func(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
# 2.2

import numpy as np
from scipy.optimize import curve_fit

b0 = 0.5
b1 = 0.1
b2 = 0.1

points_number = 150

def func(x, b0, b1, b2):
    return b0 + b1 * x + b2 * x**2

beta = (b0, b1, b2)

xdata = np.linspace(0, 5, points_number)

y = func(xdata, *beta)
ydata = y + 0.05 * np.random.randn(len(xdata))

beta_opt, beta_cov = curve_fit(func, xdata, ydata)

lin_dev = sum(beta_cov[0])
print("Linear devation: ", lin_dev)

residuals = ydata - func(xdata, *beta_opt)
fres = sum(residuals**2)
print("Quadratic devation: ", lin_dev)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, func(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
# 2.3

import numpy as np
from scipy.optimize import curve_fit

b0 = 6
b1 = 1

points_number = 150

def func(x, b0, b1):
    return b0 + b1 * np.log(x)

beta = (b0, b1)

xdata = np.linspace(1, 5, points_number)

y = func(xdata, *beta)
ydata = y + 0.05 * np.random.randn(len(xdata))

beta_opt, beta_cov = curve_fit(func, xdata, ydata)

lin_dev = sum(beta_cov[0])
print("Linear devation: ", lin_dev)

residuals = ydata - func(xdata, *beta_opt)
fres = sum(residuals**2)
print("Quadratic devation: ", lin_dev)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, func(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
# 3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

dataset = pd.read_csv("https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv")
dataset.head()

plt.scatter(dataset['YearsExperience'], dataset['Salary'], color='b')
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Coeff: ", regressor.coef_)

y_pred = regressor.predict(X_test)

predict_data = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(predict_data)

predict_data.plot(kind='bar')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()

```

```python
#4

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sI
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

dataset = pd.read_csv("https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv")
dataset.head()

X = dataset[[
    'fixed acidity',
    'volatile acidity',
    'citric acid',
    'residual sugar',
    'chlorides',
    'free sulfur dioxide',
    'total sulfur dioxide',
    'density',
    'pH',
    'sulphates',
    'alcohol'
]]
y = dataset['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

coeff_dataset = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
display(coeff_dataset)

y_pred = regressor.predict(X_test)
check_base = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
display(check_base)

print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
```
