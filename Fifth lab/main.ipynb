```python
# 3

class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def print_tree(self, depth=0, arr=None):
        try:
            left_value = self.left.data
        except AttributeError:
            left_value = None

        try:
            right_value = self.right.data
        except AttributeError:
            right_value = None

        if depth == 0:
            arr = [[self.data], [left_value, right_value]]

            if left_value is not None:
                self.left.print_tree(depth + 1, arr)
            if right_value is not None:
                self.right.print_tree(depth + 1, arr)

            self.format(arr)
        else:
            try:
                arr[depth + 1].append(left_value)
                arr[depth + 1].append(right_value)
            except IndexError:
                arr.append([left_value, right_value])

            if left_value is not None:
                self.left.print_tree(depth + 1, arr)
            if right_value is not None:
                self.right.print_tree(depth + 1, arr)

    @staticmethod
    def format(arr):
        for item in arr:
            print(item)


    def add(self, child_data):
        if self.data > child_data:
            if self.left is None:
                self.left = Tree(child_data)
                return
            if self.right is None:
                self.right = Tree(child_data)
                return

            if self.left.data > child_data or self.right.data > child_data:
                if self.left.data > child_data and self.right.data > child_data:
                    if child_data < self.left.data < self.right.data:
                        self.left.add(child_data)
                        return
                    if child_data < self.right.data < self.left.data:
                        self.right.add(child_data)
                        return
                else:
                    if child_data < self.left.data:
                        self.left.add(child_data)
                        return
                    if child_data < self.right.data:
                        self.right.add(child_data)
                        return
            else:
                if self.left.data < self.right.data:
                    tmp = self.left.data
                    self.left.data = child_data
                    self.left.add(tmp)
                    return
                else:
                    tmp = self.right.data
                    self.right.data = child_data
                    self.right.add(tmp)
                    return


tree = Tree(100)
tree.add(90)
tree.add(20)
tree.add(32)
tree.add(42)
tree.add(8)
tree.add(88)
tree.add(91)
tree.add(95)
tree.add(5)

tree.print_tree()
```

```python
# 4

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import numpy as np

X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    target,
    test_size = 0.20
)

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

tree.plot_tree(classifier)
y_pred = classifier.predict(X_test)
print(y_test, y_pred, end="\n\n")

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# 5

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn import tree
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv"

dataset = pd.read_csv(url)

display(dataset.head())

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 0
)

regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)

tree.plot_tree(regressor)

y_pred = regressor.predict(X_test)

result = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
display(result)

print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
```
